```python
import numpy as np
import matplotlib.pyplot as plt
#The number of iterations to read
r = 18
performance = []
for i in range(r):
    performance.append(np.load('models_Three/' + str('iteration_' + str(i) + '_performance.npy')))
performance = np.array(performance)
colors = 'rgbcy'
iterations = [0, 3, 6, 9, 12]
```

```python
#Plot of cv accuracy against steps, fixed iterations
#This plot will be empty if in round 3, where for each data set we only attempt one model.
for i in range(5):
    plt.plot(performance[iterations[i]], colors[i], label='iteration ' + str(iterations[i]))
plt.xlabel('steps')
plt.ylabel('cross-validation-accuracy')
plt.legend()
plt.title('policy performance at a fixed iteration')
plt.show()
```

```python
#Plot of cv accuracy against iterations, fixed steps
plt.clf()
performance = []
for i in range(r):
    performance.append(np.load('models_Three/' + str('iteration_' + str(i) + '_performance.npy')))
performance = np.array(performance)
colors = 'rgbcy'
steps = [0]
for i in range(1):
    plt.plot(performance[:,steps[i]], colors[i], label='step ' + str(steps[i]))
plt.xlabel('iterations')
plt.ylabel('cv accuracy')
plt.legend()
plt.title('policy performance at a fixed step')
plt.show()
```
